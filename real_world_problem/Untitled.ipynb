```python
# Importing base libries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# importing the dataset
data_frame = pd.read_csv('adult.csv')
# age, is_employed_by_the_state_gov, user_id, eduction_level, no_of_project, marriatal_staus, job_description, family_relation, skin_color, gender, ====, country, y_value
data_frame.head()
```

```python
# Printing the columns for this data set
data_frame.columns
# dropping the unnecery colums for a more accurate model
data_frame.drop([' 2174', ' 0', ' 40'], axis="columns", inplace=True)
data_frame.head()
```

```python
# creating new heading columns
data_frame.columns = ['Age', 'Type_of_ownwer', 'id', 'education', 'No_of_project_done', 'Marital_status', 'Job_designition', 'Family_relation', 'Race', 'Gender', 'Country', 'Salary']
data_frame.head()
```

```python
# Cheaking the shape of this dataset
data_frame.shape
```

```python
# getting more infomation about this dataset provided by padas utils
data_frame.info()
```

```python
# cheking out the description of the data_set
data_frame.describe()
```

```python
# chrecking if their are null value
data_frame.isnull().sum()
```

```python
# Getting the features and labels
X = data_frame.drop(['Age', 'id', 'Family_relation', 'Country', 'Salary'], axis = 1)
y = data_frame.drop(['Age', 'Type_of_ownwer', 'id', 'education', 'No_of_project_done', 'Marital_status', 'Job_designition', 'Family_relation', 'Race', 'Gender', 'Country'], axis = 1)
X.head()
print(y.shape)
```

```python
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
```

```python
# Label Encoding is carried out here
X['Type_of_ownwer'] = label_encoder.fit_transform(X['Type_of_ownwer'])

X['education'] = label_encoder.fit_transform(X['education'])

X['Marital_status'] = label_encoder.fit_transform(X['Marital_status'])

X['Job_designition'] = label_encoder.fit_transform(X['Job_designition'])

X['Race'] = label_encoder.fit_transform(X['Race'])

X['Gender'] = label_encoder.fit_transform(X['Gender'])

y['Salary'] = label_encoder.fit_transform(y['Salary'])

```

```python
# Preparng the dataset and spliting it
from sklearn.model_selection import train_test_split

X_train, X_test,  y_train, y_test = train_test_split(X, y, test_size=0.2)
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

# *Now kicking of the training process*

```python
# 1. Working with linear regression
# Training the Multiple Linear Regression model on the Training set
from sklearn.linear_model import LinearRegression

regressor = LinearRegression()
regressor.fit(X_train, y_train)

# Predicting the Test set results
y_pred = regressor.predict(X_test)
print(y_pred)
```

```python
print(y_test[:30], "Y test")
print(y_pred[:30], "Y predict")
```

```python
# Training the Decision Tree Regression model on the whole dataset
from sklearn.tree import DecisionTreeRegressor
regressor = DecisionTreeRegressor(random_state = 0)
regressor.fit(X_train, y_train)


# Predicting a new result
y_pred = regressor.predict(X_test)
print(y_test[:30], "Y test")
print(y_pred[:30], "Y predict")
```

```python
# Training the Random Forest Regression model on the whole dataset
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import accuracy_score
regressor = RandomForestRegressor(n_estimators = 10, random_state = 0)
regressor.fit(X_train, y_train)

# Predicting a new result
y_pred = regressor.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
```
